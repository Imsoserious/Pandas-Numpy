```python
import numpy as np
import matplotlib.pyplot as plt
```

# np.interp : 1차원 선형 보간
[numpy link](https://docs.scipy.org/doc/numpy/reference/generated/numpy.interp.html)
- numpy.interp(x, xp, fp, left=None, right=None, period=None)
- x: 보간할 x 좌표
- xp: x 좌표 데이터
- fp : y 좌표 데이터
- left : Value to return for x < xp[0], default is fp[0].
- light : Value to return for x > xp[-1], default is fp[-1].
    - (left, right는 x 범위 외에 보간에 대한 설정)
- period : 주기함수로 보간하고 싶을 때 지정(지정하면left, right는 무시됨)

**Example**

```python
xp = [1,2,3,4,6]
fp =[1,4,9,16,36]
x = [1,2,3,4,5,6]
```

일반 선형 보간

```python
plt.plot(xp, fp, label='Origin')
plt.plot(x,np.interp(x, xp, fp), '*', markersize=10, label='Interp')
plt.legend()
plt.show()
```

주기 적용한 선형 보간

```python
plt.plot(xp, fp, label='Origin')
# 주기가 2이므로 y값이 9, 36을 반복함
plt.plot(x,np.interp(x, xp, fp, period=2), '-*', markersize=10, label='Interp')
plt.legend()
plt.show()
```

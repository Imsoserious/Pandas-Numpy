# Pandas의 transform()

- DataFrame을 인자로 받아서 연산한 후 다시 DataFrame 형태로 반환하는 메소드이다.
- Groupby 뒤에 써서, 그룹별로 새로운 변수를 생성할 때 유용하게 쓸 수 있다.
- transform() 안에는 연산할 함수를 자유롭게 넣을 수 있다.

```python
import numpy as np
import pandas as pd
```

```python
# Titanic 데이터로 실습
train = pd.read_csv('./train.csv')
```

```python
train.head()
```

```python
# Ticket 번호가 같은 승객들이 있다. 
# 이들을 '일행 수'로 정의하고 변수로 만들 것이다. 
train[['PassengerId','Ticket']].groupby('Ticket').count().head()
```

```python
# Ticket_count 변수 생성
# groupby 앞에는 계산되는 df의 변수 하나
# groupby 안에는 df[변수] 형태로 인자를 넣어줘야 groupby가 가능하다.
train['Ticket_count'] = train['PassengerId'].groupby(train['Ticket']).transform('count')
```

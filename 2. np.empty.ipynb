# np.empty
- 배열의 값들을 초기화하지 않고, 메모리에 있는 남아있는 값을 그대로 쓰는 배열 생성 함수

- 장점: 배열을 생성할 때 특정한 값으로 초기화하지 않으므로 시간이 단축된다.
- 단점: 초기값을 미리 알 수 없다.

**Example**

```
import numpy as np

np.empty((2,2))
```

**np.zeros와 시간 비교**

```
%time
a = np.empty((10000,10000))
```

```
%time
a = np.zeros((10000,10000))
```

- np.empty가 더 빠르다.
